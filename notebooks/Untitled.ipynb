```python
import pandas as pd
from utils import draw_plot, parse_number
```

```python
data = '../data/data2.csv'
table = pd.read_csv(data)
for col in ['Cena', 'Zmiana', 'Minimum', 'Maksimum', 'Wolumen obrotu', 'Obroty', 'Średni wolumen', 'Średnie obroty']:
    if col in table.columns:
        table[col] = table[col].apply(parse_number)
```

```python
table["Datetime"] = pd.to_datetime(table["Data"] + " " + table["Godzina"])
table.sort_values(["Ticker", "Datetime"], inplace=True)
table["Zwrot"] = table.groupby("Ticker")["Cena"].pct_change() * 100
table["Spread"] = table["Maksimum"] - table["Minimum"]
table["Wolumen"] = table["Wolumen obrotu"]
table["Godzina"] = table["Datetime"].dt.hour
table = table[['Ticker','Cena', 'Zmiana', 'Minimum', 'Maksimum', 'Wolumen obrotu', 'Obroty', 'Średni wolumen', 'Średnie obroty', 'Zwrot', 'Spread', 'Wolumen','Godzina']]
table
```

```python
draw_plot("PKOBP", table)
```

```python
tickers = table['Ticker'].unique()
for ticker in tickers:
    draw_plot(ticker, table)
```
