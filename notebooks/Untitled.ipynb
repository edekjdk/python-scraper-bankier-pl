```python
import pandas as pd
from utils import draw_plot, parse_number
```

```python
data = '../data/data2.csv'
table = pd.read_csv(data)

```

```python
table["Datetime"] = pd.to_datetime(table["Data"] + " " + table["Godzina"])
table["Datetime"] = table["Datetime"].dt.floor("12min") 
table = table.sort_values(["Ticker", "Datetime"])
table["Hour"] = table["Datetime"].dt.hour
table["Return"] = table.groupby("Ticker")["Cena"].pct_change() * 100
table["Spread"] = table["Max 1D"] - table["Min 1D"]
table["Wolumen"] = table["Wolumen obrotu"]
table["Obroty"] = table["Wartość obrotu"]
table = table.drop("System notowań", axis = 1)
table
```

```python
draw_plot("PKOBP", table)
```

```python
tickers = table['Ticker'].unique()
for ticker in tickers:
    draw_plot(ticker, table)
```

```python
table["Return_rounded"] = table.groupby("Ticker")["Cena"].pct_change() * 100

zwroty = table.pivot_table(index="Datetime", columns="Ticker", values="Return_rounded")
korelacje = zwroty.corr()
korelacje #heatmapa korelacji
zwroty
```

```python
table = table.sort_values(["Ticker", "Datetime"])
table["VolumeDelta"] = table.groupby("Ticker")["Wolumen"].diff()

wolumen_w_godzinach = table.groupby("Hour")["VolumeDelta"].sum().reset_index()
wolumen_w_godzinach #wykres liniowy
```

```python
koniec_sesji = table["Hour"].max()
table_koniec = table[table["Hour"] == koniec_sesji]
top5_wzrost = table_koniec.nlargest(5, "Return")
top5_spadek = table_koniec.nsmallest(5, "Return")
#top5_spadek - barchart jeden dla wszystkich pokazujacy spadek, to samo wzrost
top5_wzrost
```

```python
#Spread
spread_w_czasie = table.pivot_table(index="Datetime", columns="Ticker", values="Spread")
sredni_spread = table.groupby(["Ticker"])["Spread"].mean().reset_index()
sredni_spread
```
